```python
import pandas_datareader.data as web
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
ticker = "TSLA AAPL".split(" ")
start = pd.to_datetime('2020-01-02')
end = pd.to_datetime('today')
data = web.DataReader(name=ticker, data_source='yahoo', start=start, end=end)
```

```python
data.head()
```

```python
df = round(data['Adj Close'], 2)
df
```

```python
plt.figure(figsize=(10,8))
df.plot()
plt.show()
```

```python
df.index
```

```python
df.resample('M').mean()
```

```python
round(df.resample('M').mean(), 2).plot(figsize=(8,6))
```

```python
df.resample('M').mean().apply(lambda x: x/x[0]).plot(figsize=(8,6))
```

```python
df.resample('M').mean().apply(lambda x: x/x[0])
```

```python
df.resample('M').apply(lambda x: x[-1]/x[0]-1)
```

```python
df.resample('M').apply(lambda x: x[-1]/x[0]-1).plot.bar(figsize=(10,8) )
plt.show()
```

```python
df_apple = pd.DataFrame(df['AAPL'])
```

```python
df_apple['lag(t-1)'] = df['AAPL'].shift(-1)
df_apple['lead(t+1)'] = df['AAPL'].shift(+1)
df_apple.head()
```

```python
df['TSLA_MA(30)'] = df['TSLA'].rolling(window=30).mean()
df['TSLA_MA(5)'] = df['TSLA'].rolling(window=5).mean()

df.tail()
```

```python
df['TSLA_MA(5)std'] = df['TSLA'].rolling(window=30).std()
df.head
```

```python
df[['TSLA', 'TSLA_MA(30)']].plot(figsize=(12,6))
```
